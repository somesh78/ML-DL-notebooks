```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

data = pd.read_csv('/content/movie_metadata.csv')

cols_to_impute = data.columns[data.isnull().sum() > 100]

for column in cols_to_impute:
  if data[column].dtype == 'float64' or data[column].dtype == 'int64':
        data[column].fillna(data[column].median(), inplace=True)
  else:
        data[column].fillna(data[column].mode()[0], inplace=True)
data.dropna(inplace=True)

data.dropna(inplace=True)

label_encoder = LabelEncoder()
cat_cols = ['language', 'country', 'content_rating']
for col in cat_cols:
    data[col] = label_encoder.fit_transform(data[col])

one_hot_encoder = OneHotEncoder(sparse=False)
country_encoded = one_hot_encoder.fit_transform(data[['country']])
country_encoded_df = pd.DataFrame(country_encoded, columns=one_hot_encoder.get_feature_names_out(['country']))
data = pd.concat([data, country_encoded_df], axis=1)
data.drop(columns=['country'], inplace=True)

print(data)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import random

data = {'x': [random.randint(0, 200) for _ in range(0, 250)],
        'y': [random.randint(0, 200) for _ in range(0, 250)]}

df = pd.DataFrame(data)

X = df[['x']]
y = df['y']

model = LinearRegression()
model.fit(X, y)

r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.2f}")

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Linear Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

for _ in range(0,100):

  data = {'x': [random.randint(0, 100) for _ in range(0, 250)],
          'y': [random.randint(0, 50) for _ in range(0, 250)]}

  df = pd.DataFrame(data, columns=['x', 'y'])

  mean_x = df['x'].mean()
  mean_y = df['y'].mean()

  numerator = sum((df['x'] - mean_x) * (df['y'] - mean_y))
  denominator = sum((df['x'] - mean_x) ** 2)
  slope = numerator / denominator

  intercept = mean_y - slope * mean_x

  print(f"The equation of the regression line is: y = {slope:.2f}x + {intercept:.2f}")

  plt.scatter(df['x'], df['y'], color='blue')
  plt.plot(df['x'], intercept + slope * df['x'], color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
